```python
# 연령(1~100세) 데이터 result에 저장
import csv
f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data :
    if name in row[0]:
        for i in row[3:]:
            i = i.replace(',', '')
            result.append(int(i))
print(result)
```

```python
# 곡선 그래프
import matplotlib.pyplot as plt
plt.style.use('ggplot') #격자무늬
plt.rc('font', family='Malgun Gothic') #글씨 폰트설정
plt.title(name+ ' 지역의 인구 구조') #그래프 제목설정
plt.plot(result)
plt.show()
```

```python
# 수직바 그래프
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
# 수평바 그래프
import matplotlib.pyplot as plt
plt.barh(range(101), result)
plt.show()
```
