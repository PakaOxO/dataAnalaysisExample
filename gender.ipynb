```python
# 지역별 성별로 나누어 나이별 인구수
import csv
f = open('gender.csv')
data = csv.reader(f)
name = input('지역별, 인구수를 알기 위해서는 지역명(읍면동)을 입력해주세요 : ')
m = []
f = []
for row in data:
    if name in row[0]:
        for i in row[3: 104]:
            i = i.replace(',', '')
            m.append(-int(i))
        for j in row[106:]:
            j = j.replace(',', '')
            f.append(int(j))
print(m)
print(f)
```

```python
# 항아리 그래프
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('강내면 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성') # 남성 인구를 (-)음수로 두어 항아리형태로 도출
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

```python
# 지역별 인구수 성비율
import csv
f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('지역의 성비를 알기 위해서는 지역명(읍면동)을 입력해주세요 : ')
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
plt.title(name+' 지역의 남녀 성별비율')
plt.show()
```
